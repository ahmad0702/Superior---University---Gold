# Hang Man Game :

```python
import random
words = ["apple", "banana", "cherry", "date", "elderberry"]
word = random.choice(words)
guessed = ["_"] * len(word)
tries = 6
hangman = [
    r"""
    +---+
    |   |
        |
        |
        |
        |
    =========""",
    r"""
    +---+
    |   |
    O   |
        |
        |
        |
    =========""",
    r"""
    +---+
    |   |
    O   |
    |   |
        |
        |
    =========""",
    r"""
    +---+
    |   |
    O   |
   /|   |
        |
        |
    =========""",
    r"""
    +---+
    |   |
    O   |
   /|\  |
        |
        |
    =========""",
    r"""
    +---+
    |   |
    O   |
   /|\  |
   /    |
        |
    =========""",
    r"""
    +---+
    |   |
    O   |
   /|\  |
   / \  |
        |
    ========="""
]
print("Hangman Game!")
print(hangman[0])
print(" ".join(guessed))
while "_" in guessed and tries > 0:
    guess = input("Guess a letter: ").lower()
    if len(guess) != 1:
        print("Guess one letter at a time.")
    elif guess in guessed:
        print("Already guessed.")
    elif guess not in word:
        tries -= 1
        print(hangman[6-tries])
        print(f"Tries left: {tries}")
    else:
        for i, letter in enumerate(word):
            if letter == guess:
                guessed[i] = guess
    print("\n")
    print(hangman[6-tries])
    print(" ".join(guessed))
    print(f"Tries left: {tries}")
if "_" not in guessed:
    print("You won!")
else:
    print("Game over! Word: " + word)
```
